```r
library(ggplot2)

cytof_data <- read.csv("cytof_one_experiment.csv")
head(cytof_data)
colnames(cytof_data)

# 2. Choose one of the columns of the dataset and make at least two plots showing its distribution (choose from ECDF, quantile, histogram, density estimate).

cytof.gg <- ggplot(cytof_data, aes(x = TRAIL))

# ECDF
cytof.gg + stat_ecdf(geom = "step",
    color = "pink",
    linewidth = 1.2) +
  xlab("TRAIL") +
  ylab("Percentile") +
  ggtitle("ECDF Plot with ggplot2") + theme_minimal()

# density estimate
cytof.gg + geom_density(adjust = 0.5, fill = "lightblue", color = "black",
    alpha = 0.6) + geom_rug(aes(y = 0),sides = "b", position = position_jitter(height = 0)) +
    geom_vline(aes(xintercept = mean(TRAIL)), color = "red", linetype = "dashed", linewidth = 1) +
    xlab("TRAIL") +
    ylab("Density") +
    ggtitle("Density Plot of Cytof Trail with Mean Line") +theme_minimal()
```

```r
# 3. Choose any two of the columns of the dataset and make a Q-Q plot comparing the distributions.

qq.df <- as.data.frame(
  qqplot(cytof_data$ILT2, cytof_data$KIR2DL3, plot.it = FALSE)
)

ggplot(qq.df, aes(x = x, y = y)) +
  geom_point(shape = 21, fill = NA, color = "black", size = 2,
  stroke = 1, alpha = 0.6) + geom_abline(slope = 1,intercept = 0, linetype = "dashed",
  color = "red", linewidth = 1) +
  xlab("ILT2 Quantiles") +
  ylab("KIR2DL3 Quantiles") +
  ggtitle("Q-Q Plot Comparing ILT2 and KIR2DL3") +
  theme_minimal()


```
